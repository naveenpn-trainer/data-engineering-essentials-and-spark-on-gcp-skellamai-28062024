# Functions

> Functions are a set of actions that we group together, and give a name to it.

`
def function_name(param_01,param_02):
  # Function Body
  # return value # Optional
`

```python
def calculate_percentage(obtained_marks,total_marks):
  print(f"Calculating percentage for marks: {obtained_marks} with total marks: {total_marks}")
  return ((obtained_marks * 100)/ total_marks)


calculate_percentage(70,100)
```

## Position Arguments

```python
calculate_percentage(70,100)
```

## Named Arguments

```python
calculate_percentage(total_marks = 100, obtained_marks= 60)
```

## Default Arguments

```python
def calculate_percentage(obtained_marks,total_marks=100):
  print(f"Calculating percentage for marks: {obtained_marks} with total marks: {total_marks}")
  return ((obtained_marks * 100)/ total_marks)


calculate_percentage(70,100)
```

```python
calculate_percentage(70)
calculate_percentage(80)
calculate_percentage(90,200)
```

# Lambda Functions

> A Lambda function is a small anonymous function defined using the lambda keyword

*syntax*

`
lambda arguments : expression
`

```python
def add(x,y):
  return x+y

add(1,3)
```

```python
add = lambda x,y:x+y
add(9,10)
```

```python
2+2
```

```python
def multiply_by(number):
  return lambda x:x*number


double = multiply_by(5)
double(2)
```
